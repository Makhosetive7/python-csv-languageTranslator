```python
import pandas as pd
from googletrans import Translator

#initializing translator object
Translator = Translator()
```

```python

#loading the csv dataset
df = pd.read_csv('./ViMedical_Disease.csv')
outputFile = "translated_ViMedical_Disease.csv"
```

```python
#Function to translate the the whole dataset
def translate_dataset(text, src_lang='vi', dest_lang='en'):
    try:
        translated = Translator.translate(text, src=src_lang, dest=dest_lang)
        return translated.text
    except Exception as e:
        print(f"Error translating the dataset: {e}")
        return text
    
#Translating the dataset
df['Disease'] = df['Disease'].apply(translate_dataset)
df['Question'] = df['Question'].apply(translate_dataset)

#Writing the translated dataset to a new csv file
df.to_csv(outputFile, index=False)
print(f"Translated dataset saved to {outputFile}")
```
