```python
import pandas as pd
import googletrans
from googletrans import Translator
import time
```

```python
# Loading the txt dataset
inputFile = ("./data/input/ViMedical_Disease.csv")  # Replace with your input file path
df = pd.read_csv(inputFile)  # Use '\t' for tab delimiter or ',' for comma

df.head(3)
```

```python
# Instantiate the translator
translator = Translator()

#function to translate a text using Google Translate API
def translate_dataset(text, sourceLanguage='vi', destinationLanguage='en'):
    translator = Translator()
    try:
        translated = translator.translate(text, src=sourceLanguage, dest=destinationLanguage)
        return translated.text
    except Exception as e:
        print(f"Error translating the dataset: {e}")
        return text

# Split the DataFrame into smaller chunks
chunk_size = 50  # Adjust this number based on your needs
chunks = [df[i:i + chunk_size] for i in range(0, df.shape[0], chunk_size)]

# Translate each chunk and combine them back
translated_chunks = []
for chunk in chunks:
    chunk.loc[:, 'Disease'] = chunk['Disease'].apply(lambda x: translate_dataset(x))
    chunk.loc[:, 'Question'] = chunk['Question'].apply(lambda x: translate_dataset(x))
    translated_chunks.append(chunk)
    time.sleep(10)

translated_df = pd.concat(translated_chunks)

# Check the translated data
print(translated_df.head(3))

# Writing the translated dataset to a new CSV file
outputFile = "./data/output/translated_ViMedical_Disease.csv"
translated_df.to_csv(outputFile, index=False)
print(f"Translated dataset saved to {outputFile}")
```
