```python
import pandas as pd
import googletrans
from googletrans import Translator
```

```python

#loading the csv dataset
inputFile = "./ViMedical_Disease.csv"
df = pd.read_csv(inputFile)


df.head(20)
```

```python
#Function to translate the the whole dataset
def translate_dataset(text, sourceLanguage='vi', destinationLanguage='en'):
     # Create a translator object
    translator = Translator()
    try:
        translated = translator.translate(text, src=sourceLanguage, dest=destinationLanguage)
        return translated.text
    except Exception as e:
        print(f"Error translating the dataset: {e}")
        return text
    
#Translating the dataset
df['Disease'] = df['Disease'].apply(lambda x: translate_dataset(x))
df['Question'] = df['Question'].apply(lambda x: translate_dataset(x))

#Writing the translated dataset to a new csv file
outputFile = "translated_ViMedical_Disease.csv"
df.to_csv(outputFile, index=False)
print(f"Translated dataset saved to {outputFile}")
```
